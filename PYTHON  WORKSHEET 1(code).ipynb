# 11. Write a python program to find the factorial of a number.

```python
def factorial(num):
    fact=1;
    i=1
    while i<=num:
        fact=fact*i
        i=i+1
    return fact
    
num=int(input("Enter a number.."));
result=factorial(num)
print("factorial of the number=",result)
factorial(num)
```

# 14. Write a Python program to get the third side of right-angled triangle from two given sides

```python
import math
def base(height,hypotenus):
    base=math.sqrt((hypotenus**2)-(height**2))
    print("The base of the Triangle is=",base)
def height(base,hypotenus):
    height=math.sqrt((hypotenus**2)-(base**2))
    print("The Height of the Triangle is=",height)
def hypotenus(height,base):
    hypotenus=math.sqrt((height**2)+(base**2))
    print("The Hypotenus of the Tringle is=",hypotenus)
    
p=input("Type one of height,base or hypotenus for which side required of the triangle")
if p=='height':
    base=int(input("Please input value of base of the Triangle"))
    hypotenus=int(input("please input value of hypotenus of the Triangle"))
    height(base,hypotenus)
if p=='base':
    height=int(input("Please input value of height of the Triangle"))
    hypotenus=int(input("please input value of hypotenus of the Triangle"))
    base(height,hypotenus)
if p=='hypotenus':
    base=int(input("Please input value of base of the Triangle"))
    height=int(input("Please input value of height of the Triangle"))
    hypotenus(height,base)
```
